```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
H=nx.DiGraph()
H.add_edge('A','B')
H.add_edge('B','C')
H.add_edge('B','D')
H.add_edge('B','E')
H.add_edge('D','E')
H.add_edge('D','C')
H.add_edge('D','C')
nx.draw_networkx(H)
plt.show()
```

```python
G = nx.DiGraph ()
G.add_edge('A', 'B');
G.add_edge('A', 'C');
G.add_edge('B', 'D');
G.add_edge('B', 'E');
G.add_edge('D', 'E');
nx.draw_networkx(G)
plt.show()
```

```python
H = nx.DiGraph ()
H.add_edge('A', 'B');
H.add_edge('A', 'C');
H.add_edge('B', 'D');
H.add_edge('B', 'E');
H.add_edge('D', 'E');
H.add_edge('D', 'C');
H.add_edge('D', 'A');
nx.draw_networkx(H)
plt.show()
```
