## Import required library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import seaborn as sns
```

## Practice 1:  Descriptive Statistics with python 

```python
# read csv files
df=pd.read_csv('haberman.csv')
df.head()
```

```python

# Figure 1
df.boxplot()
plt.show()
```

```python
# Figure 2
plt.boxplot(df.age)
plt.title('Boxplot for Afe of Breast Cancer Patient')
```

```python
## Figure 3
### Histrogram
plt.hist(df.age,bins=15)
plt.xlabel('Age of Patients')
plt.ylabel('Histogram of Age of Breast Cancer Patients')
plt.show()

```

```python
## Figure 4
plt.boxplot(df.axil_nodes)
plt.title('Boxplot for Axillary Nodes of Breast Cancer Patients')
 
```

```python
# Figure 13
plt.scatter(df['operation_year'],df['axil_nodes'],color='r')
plt.xlabel('Operation Year')
plt.ylabel('Axillary Nodes')
plt.title('Operation Year Vs Axil Nodes')
plt.show()
```
